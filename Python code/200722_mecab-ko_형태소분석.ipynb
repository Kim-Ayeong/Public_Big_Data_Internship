```python
from eunjeon  import Mecab
```

```python
tagger = Mecab()
```

```python
txt ='나는 하늘을 나는 비행기를 보았습니다.'
```

```python
# 문장을 형태소 단위 분리
tagger.morphs(txt)
```

```python
# 명사 추출
tagger.nouns(txt)
```

```python
# 문장을 형태소단위 분리후 형태소 마다 품사를 분석
tagger.pos(txt)
```
